```python
import pandas as pd
import numpy as np
import random
import string
np.random.seed(56)
```

```python
def score():
    return np.random.randint(0, 100)
```

```python
def id():
    return ''.join(random.choice(string.ascii_letters) for i in range(5))
```

```python
df = pd.DataFrame({"국어":score(), "영어":score(), "수학":score(), "사회":score()}, index=[id()])
for i in range(1,10):
    df.loc[id()] = {"국어":score(), "영어":score(), "수학":score(), "사회":score()}
df
```
